```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
chess_2D=cv2.imread('C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/flat_chessboard.png')
display=cv2.cvtColor(chess_2D,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(display)
```

```python
gray=cv2.cvtColor(chess_2D,cv2.COLOR_BGR2GRAY)
```

```python
gray=np.float32(gray)
```

```python
dst=cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst=cv2.dilate(dst,None)
```

```python
display[dst>0.01*dst.max()]=[255,0,0]
plt.imshow(display)
```

```python
chess_3D=cv2.imread('C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
display=cv2.cvtColor(chess_3D,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(display)
```

```python
gray=cv2.cvtColor(chess_3D,cv2.COLOR_BGR2GRAY)
```

```python
gray=np.float32(gray)
```

```python
dst=cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)
```

```python
display[dst>0.01*dst.max()]=[255,0,0]
```

```python
plt.imshow(display)
```

```python
fig=plt.figure(figsize=(12,10))
ax=fig.add_subplot(111)
ax.imshow(display)
```

```python
corners=cv2.goodFeaturesToTrack(gray,100,0.01,10)
```

```python
corners=np.int0(corners)
```

```python
for i in corners:
    x,y=i.ravel()
    cv2.circle(display,(x,y),5,(255,0,0),-1)
```

```python
plt.imshow(display)
```
